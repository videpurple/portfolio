````python
# 추가 전처리
# 거래 데이터에 Year-Month 컬럼 추가 및 프로모션 기간 여부 컬럼 추가
## Year-Month 컬럼 추가
ce['order_date'] = pd.to_datetime(ce['order_date'])
ce['Year_Month'] = ce['order_date'].dt.strftime('%Y-%m')
## 프로모션 여부 컬럼 추가 (special sale 기간이면 1 아니면 0)
conditions = (
    (ce['order_date'].between('2015-07-18 00:00:00', '2015-07-19 23:59:59')),
    (ce['order_date'].between('2015-08-15 00:00:00', '2015-08-17 23:59:59')),
    (ce['order_date'].between('2015-08-28 00:00:00', '2015-08-30 23:59:59')),
    (ce['order_date'].between('2015-10-15 00:00:00', '2015-10-17 23:59:59')),
    (ce['order_date'].between('2015-11-07 00:00:00', '2015-11-14 23:59:59')),
    (ce['order_date'].between('2015-12-25 00:00:00', '2016-01-03 23:59:59')),
    (ce['order_date'].between('2016-01-20 00:00:00', '2016-01-22 23:59:59')),
    (ce['order_date'].between('2016-02-01 00:00:00', '2016-02-02 23:59:59')),
    (ce['order_date'].between('2016-02-14 00:00:00', '2016-02-15 23:59:59')),
    (ce['order_date'].between('2016-02-20 00:00:00', '2016-02-21 23:59:59')),
    (ce['order_date'].between('2016-03-07 00:00:00', '2016-03-09 23:59:59')),
    (ce['order_date'].between('2016-05-25 00:00:00', '2016-05-27 23:59:59'))
)
choices = [1] * len(conditions)
ce['Special_Sale'] = np.select(conditions, choices, default=0)

# 광고 및 투자내역 데이터 컬럼명 공백 제거 및 Year_Month 컬럼 추가
mi.rename(columns = {' Affiliates' : 'Affiliates'}, inplace = True)
mi['Year_Month'] = mi['Year'].apply(str) + '-' + mi['Month'].apply(str)
mi['Year_Month'] = pd.to_datetime(mi['Year_Month'], format = '%Y-%m')
mi['Year_Month'] = mi['Year_Month'].dt.strftime('%Y-%m')

# 고객 충성도, 주가지수 데이터 형태 변경 및 'Year_Month'컬럼 수정
ns.set_index(keys = 'index', inplace = True)
ns = ns.transpose()
ns.reset_index(inplace = True)
ns.rename(columns = {'index' : 'ym'}, inplace = True)
ns[['Month', 'Year']] = ns['ym'].str.split("'", expand = True)
ns['Year'] = '20' + ns['Year']
ns['Year_Month'] = ns['Year'] + "-" + ns['Month']
ns['Year_Month'] = pd.to_datetime(ns['Year_Month'])
ns['Year_Month'] = ns['Year_Month'].dt.strftime('%Y-%m')
````

#### 1. 연간 총수익 및 거래 현황
---

#### 재구매율 확인을 위한 코호트 분석

---
- 우리 서비스를 이용한 전체 고객(1,253,494명) 중 약 14%의 고객(172,107명)이 2회 이상 구매를 하였습니다.
- 유지율을 확인한 결과 자사의 마케팅은 신규 고객 유치에는 매우 효과적인 마케팅이 진행되고 있습니다.
- 전자 제품의 특성상 이용 주기가 길지만 게임 등 다양한 Product 판매하기도 하여 2회이상 구매한 고객 중 구매주기가 30일 ~ 40일 이내로 나타났습니다.


````python
# 코호트 분석을 위한 요약 데이터프레임 작성
stat_df = pd.DataFrame(ce.groupby('cust_id').agg({'order_date' : ['min', 'max'], 'order_id' : 'nunique'}))
stat_df.columns = ['FirstOrder', 'LastOrder', 'OrdersCount']
stat_df.reset_index(inplace = True)

# 고객별 구매금액 구한 후 stat_df와 합치기
sales = pd.DataFrame(ce.groupby(['cust_id'])['gmv'].sum().reset_index(name = 'TotalGMV'))
stat_df = pd.merge(left = stat_df, right = sales, how = 'left', on = 'cust_id')
````

````python
# 2회 이상 구매한 고객 확인
stat_copy = stat_df.copy()
stat_copy = stat_copy[stat_copy['OrdersCount'] >= 2]

stat_copy['cust_id'].value_counts().sum()
````

재구매한 고객들의 거래 주기 계산

- 분석대상 : 2015년 첫거래일 이후 1일이 지난 시점부터 거래 기록이 있는 고객 (172,107명 중 139,261명이 해당. 전체 재구매 고객 중 약 81%)
- 거래 주기 = (마지막거래일 - 첫거래일) / (거래수 - 1)

````python
# 거래주기 확인
stat_copy['diff'] = (stat_copy['LastOrder'] - stat_copy['FirstOrder'])
stat_copy['cycle'] = stat_copy['diff'] / (stat_copy['OrdersCount'] - 1)
stat_copy['diff'] = stat_copy['diff'].dt.days
stat_copy['cycle'] = stat_copy['cycle'].dt.days
stat_copy.sort_values(by = 'cycle', inplace = True)

stat_copy = stat_copy[stat_copy['diff'] > 0]

stat_copy['cycle'].describe()
````

````python
# 거래주기 분포 시각화
sns.distplot(stat_copy['cycle']);
````

리텐션 분석

- 재구매를 진행한 고객을 대상으로 구매주기를 구한 결과 평균값 31일, 중앙값 37일로 나타났습니다.
- 리텐션 분석 방법으로는 전자제품의 사용 주기가 긴 것을 감안하여 범위 리텐션으로 진행하였습니다.
- 광고 및 투자내역 데이터, 고객 충성도 및 주가지수 데이터 등 다양한 데이터들이 월별로 나타나 있음을 감안하여 기간을 묶는 기준도 해당 기준에 맞추어 1달(30일)을 기준일로 정하였습니다.

````python
# 코호트 분석 데이터 만들기 위한 날짜 데이터 변환
from datetime import datetime, timedelta

temp = pd.merge(left = ce, right = stat_df, how = 'inner', on = 'cust_id')
temp = temp[['cust_id', 'order_date', 'FirstOrder']]

temp['FirstOrderMonth'] = temp['FirstOrder'].dt.strftime('%Y-%m-01')
temp['FirstOrderMonth'] = temp['FirstOrderMonth'].str.slice(start = 0, stop = 7)
temp['FirstOrder'] = temp['FirstOrder'].dt.strftime('%Y-%m-%d')
temp['order_date'] = temp['order_date'].dt.strftime('%Y-%m-%d')

# temp['FirstOrderMonth'] = pd.to_datetime(temp['FirstOrderMonth'])
temp['FirstOrder'] = pd.to_datetime(temp['FirstOrder'])
temp['order_date'] = pd.to_datetime(temp['order_date'])
````

````python
# 반복문 이용하여 코호트 차트 작성
cohort = temp.copy()
ci_0 = pd.DataFrame(cohort.groupby('FirstOrderMonth')['cust_id'].nunique())
ci_0.rename(columns = {'cust_id' : 'Month 0'}, inplace = True)

cohort = ci_0

for i in range(0, 12) :
  j = int(i)
  i = temp[(temp['order_date'] >= temp['FirstOrder'] + timedelta(days= 1 + (30 * j))) & (temp['order_date'] <= temp['FirstOrder'] + timedelta(days=30 * (j + 1)))]
  i = pd.DataFrame(i.groupby('FirstOrderMonth')['cust_id'].nunique())
  i.rename(columns = {'cust_id' : f'Month {1*(j+1)}'}, inplace = True)

  cohort = pd.concat([cohort, i], axis = 1)

  # day0을 제외한 값은 day0의 사람들 중 재구매한 사람들의 비율을 표시
  cohort[f'Month {1*(j+1)}'] = round(cohort[f'Month {1*(j+1)}'] / cohort['Month 0'] * 100, 2)


cohort = cohort.iloc[2:14]
cohort.fillna(0, inplace = True)

# 코호트 분석 시각화
# cohort_cust = cohort.style.background_gradient(cmap='Reds', subset=['Month 0'])

cohort_chart = cohort.style.background_gradient(subset = ['Month 1', 'Month 2', 'Month 3', 'Month 4', 'Month 5', 'Month 6', \
                                                          'Month 7', 'Month 8', 'Month 9', 'Month 10', 'Month 11', 'Month 12'],
                                                 axis = None)

cohort_chart
````

- 총 수익, 구매 고객 수, 주문 수와 상관관계가 적은 광고 채널의 투자 금액을 상관 관계가 높은 광고 채널에 투자하여 수익을 높입니다.
- 아래 머신러닝을 통해 상관관계가 높았던 스폰서십, 온라인 마케팅, 제휴 마케팅(Affiliates)의 평균 투자금액으로 투자했을 때 2016년 7월의 예상 수익을 구해보았습니다.

````python
# 머신러닝을 위한 데이터 준비

consumer_new = ce[['cust_id', 'product_analytic_super_category', 'product_analytic_category', 'product_analytic_sub_category', 'product_analytic_vertical', 'gmv', 'Year_Month']]
media_new = mi.drop(['Year', 'Month'], axis = 'columns')
consumer_media = pd.merge(consumer_new, media_new, on='Year_Month', how='left')

gmv_monthly = consumer_media.groupby('Year_Month')['gmv'].sum().reset_index()
media_investment_monthly = media_new[['Year_Month', 'TV', 'Digital', 'Sponsorship', 'Content Marketing', 'Online marketing', 'Affiliates', 'SEM', 'Radio', 'Other']]
````

````python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

# 데이터 준비
combined_data = pd.merge(gmv_monthly, media_investment_monthly, on='Year_Month', how='inner')

# 상관관계 높은 미디어 투자 유형 선택
X = combined_data[['Sponsorship', 'Online marketing', 'Affiliates']]
y = combined_data['gmv']

# 데이터를 훈련 세트와 테스트 세트로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 그래디언트 부스팅 회귀 모델 생성 및 훈련
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# 과거 미디어 투자 데이터의 평균을 계산
average_investments = media_investment_monthly[['Sponsorship', 'Online marketing', 'Affiliates']].mean()

# 총 평균 투자액 계산
total_average_investment = average_investments.sum()

# 예측을 위해 과거 평균 투자액 사용
predicted_gmv_july_2016 = model.predict([[
    average_investments['Sponsorship'],
    average_investments['Online marketing'],
    average_investments['Affiliates']
]])
print(f"Predicted GMV for July 2016 based on average investments: {predicted_gmv_july_2016[0]:.2f}")
print(f"Total average investment: {total_average_investment:.2f}")
````

````python
# GradientBoostingRegressor 정확도 계산
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 테스트 데이터에 대한 예측 수행
y_pred = model.predict(X_test)

# 성능 평가
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# 결과 출력
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")
````

예상 수익은 327,122,701.66 이 나왔습니다.

이는 평균 총 광고 투자 비용(70.55)을 투자했을 때 발생하는 월 평균 총 GMV(311,089,003.21)과 비교하여
더 적은 투자 비용으로 1.05배 높은 수익이 나올 것으로 예상됩니다.
이와 같이 수익이 잘 나올 것이라 예상되는 채널을 선택, 집중하여 총 수익을 올릴 수 있도록 반영할 수 있습니다.
