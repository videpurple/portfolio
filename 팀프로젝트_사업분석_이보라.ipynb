````python
# ë¨¸ì‹ ëŸ¬ë‹ì„ ìœ„í•œ ë°ì´í„° ì¤€ë¹„

consumer_new = ce[['cust_id', 'product_analytic_super_category', 'product_analytic_category', 'product_analytic_sub_category', 'product_analytic_vertical', 'gmv', 'Year_Month']]
media_new = mi.drop(['Year', 'Month'], axis = 'columns')
consumer_media = pd.merge(consumer_new, media_new, on='Year_Month', how='left')

gmv_monthly = consumer_media.groupby('Year_Month')['gmv'].sum().reset_index()
media_investment_monthly = media_new[['Year_Month', 'TV', 'Digital', 'Sponsorship', 'Content Marketing', 'Online marketing', 'Affiliates', 'SEM', 'Radio', 'Other']]
````

````python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split

# ë°ì´í„° ì¤€ë¹„
combined_data = pd.merge(gmv_monthly, media_investment_monthly, on='Year_Month', how='inner')

# ìƒê´€ê´€ê³„ ë†’ì€ ë¯¸ë””ì–´ íˆ¬ì ìœ í˜• ì„ íƒ
X = combined_data[['Sponsorship', 'Online marketing', 'Affiliates']]
y = combined_data['gmv']

# ë°ì´í„°ë¥¼ í›ˆë ¨ ì„¸íŠ¸ì™€ í…ŒìŠ¤íŠ¸ ì„¸íŠ¸ë¡œ ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# ê·¸ë˜ë””ì–¸íŠ¸ ë¶€ìŠ¤íŒ… íšŒê·€ ëª¨ë¸ ìƒì„± ë° í›ˆë ¨
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# ê³¼ê±° ë¯¸ë””ì–´ íˆ¬ì ë°ì´í„°ì˜ í‰ê· ì„ ê³„ì‚°
average_investments = media_investment_monthly[['Sponsorship', 'Online marketing', 'Affiliates']].mean()

# ì´ í‰ê·  íˆ¬ìì•¡ ê³„ì‚°
total_average_investment = average_investments.sum()

# ì˜ˆì¸¡ì„ ìœ„í•´ ê³¼ê±° í‰ê·  íˆ¬ìì•¡ ì‚¬ìš©
predicted_gmv_july_2016 = model.predict([[
    average_investments['Sponsorship'],
    average_investments['Online marketing'],
    average_investments['Affiliates']
]])
print(f"Predicted GMV for July 2016 based on average investments: {predicted_gmv_july_2016[0]:.2f}")
print(f"Total average investment: {total_average_investment:.2f}")
````

````python
# GradientBoostingRegressor ì •í™•ë„ ê³„ì‚°
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡ ìˆ˜í–‰
y_pred = model.predict(X_test)

# ì„±ëŠ¥ í‰ê°€
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# ê²°ê³¼ ì¶œë ¥
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"RÂ² Score: {r2:.2f}")
````
